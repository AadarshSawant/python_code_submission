```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt

nltk.download('vader_lexicon')

# Initialize VADER
sia = SentimentIntensityAnalyzer()

df = pd.read_csv('covid19_tweets.csv')
texts = df['text'].astype(str)

total_pos = 0
total_neg = 0
total_nue = 0
total_comp = 0

for text in texts:
    scores = sia.polarity_scores(text)
    total_pos += scores['pos']
    total_neg += scores['neg']
    total_nue += scores['neu']
    total_comp += scores['compound']

categories = ['POsitive', 'Negative','Nuetral','Coumpound']
values = [total_pos,total_neg,total_nue,total_comp]


plt.bar(categories,values,color=['green','red','grey','blue'])
plt.title('COVID-19 Tweets Sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Number of Tweets')
plt.show()
```
